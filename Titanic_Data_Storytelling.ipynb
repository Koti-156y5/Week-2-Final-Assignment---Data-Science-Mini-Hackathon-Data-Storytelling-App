The Titanic dataset contains information about passengers aboard the RMS Titanic. It includes details such as age, gender, passenger class, fare paid, and survival status. The objective is to analyze the data and identify factors that influenced passenger survival.

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Observations:
The dataset contains 891 passengers.
Some missing values exist in Age, Cabin, and Embarked.
Passenger ages vary widely.
Fare values show significant variation.

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df.isnull().sum()
```

Missing values in Age and Embarked were handled using mean and mode imputation.

```python
import plotly.express as px

fig = px.histogram(df, x='Survived',
                   title='Passenger Survival Count')
fig.show()
```

```python
fig = px.pie(df,
             names='Sex',
             title='Gender Distribution')
fig.show()
```

```python
fig = px.histogram(df,
                   x='Age',
                   title='Age Distribution')
fig.show()
```

```python
fig = px.histogram(df,
                   x='Pclass',
                   title='Passenger Class Distribution')
fig.show()
```

```python
fig = px.scatter(df,
                 x='Age',
                 y='Fare',
                 color='Survived',
                 title='Age vs Fare')
fig.show()
```

Key Findings:
Female passengers had a higher survival rate than male passengers.
First-class passengers were more likely to survive.
Most passengers were between 20 and 40 years old.
Higher fare-paying passengers generally had better survival chances.
Passenger class significantly influenced survival outcomes.

Conclusion:
The analysis shows that gender, passenger class, and fare played important roles in determining survival during the Titanic disaster. Female passengers and first-class travelers had a greater probability of survival. Data storytelling helps transform raw data into meaningful insights that can support decision-making and understanding of historical events.
